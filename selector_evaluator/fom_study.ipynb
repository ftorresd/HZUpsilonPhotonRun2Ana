Function that calculates the weight for each MC dataset

```python
def get_xsec(dataset):
  # REF: https://docs.google.com/spreadsheets/d/1zvvedRr1K4NFylNqxNdkWbFgoMPvh9E9BrqxywBpu-Y/edit?usp=sharing
  # define total xsecs 
  z_total_xsec = 5.7095E+04
  higgs_ggH_xsec = 4.8580E+01
  higgs_WpH_xsec = 8.4000E-01
  higgs_WmH_xsec = 5.3280E-01
  higgs_ZH_xsec = 8.8390E-01
  higgs_VBFH_xsec = 3.7820E+00
  higgs_ttH_xsec = 5.0710E-01
  higgs_total_xsec = 5.5710E+01


  # define BRs
  z_1S_br = 4.8000E-08
  z_2S_br = 2.4400E-08
  z_3S_br = 1.8800E-08
  higgs_1S_br = 5.2200E-09
  higgs_2S_br = 1.4200E-09
  higgs_3S_br = 9.1000E-10
  mumugamma_fsr_br = 1.63E-6
  dalitz_decay_br = 3.8300E-05 
  upsilon_1S_to_mumu = 2.4800E-02
  upsilon_2S_to_mumu = 1.9300E-02
  upsilon_3S_to_mumu = 2.1800E-02

  effective_xsec = None
  # Z
  if dataset["dataset_name"] == "ZToUpsilon1SGamma":
    effective_xsec = z_total_xsec*z_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "ZToUpsilon2SGamma":
    effective_xsec = z_total_xsec*z_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "ZToUpsilon3SGamma":
    effective_xsec = z_total_xsec*z_3S_br*upsilon_3S_to_mumu

  # Higgs - ggH
  if dataset["dataset_name"] == "ggH_HToUps1SG":
    effective_xsec = higgs_ggH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "ggH_HToUps2SG":
    effective_xsec = higgs_ggH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "ggH_HToUps3SG":
    effective_xsec = higgs_ggH_xsec*higgs_3S_br*upsilon_3S_to_mumu

  # Higgs - WpH
  if dataset["dataset_name"] == "WpH_HToUps1SG":
    effective_xsec = higgs_WpH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "WpH_HToUps2SG":
    effective_xsec = higgs_WpH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "WpH_HToUps3SG":
    effective_xsec = higgs_WpH_xsec*higgs_3S_br*upsilon_3S_to_mumu

  # Higgs - WpH
  if dataset["dataset_name"] == "WmH_HToUps1SG":
    effective_xsec = higgs_WmH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "WmH_HToUps2SG":
    effective_xsec = higgs_WmH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "WmH_HToUps3SG":
    effective_xsec = higgs_WmH_xsec*higgs_3S_br*upsilon_3S_to_mumu

  # Higgs - ZH
  if dataset["dataset_name"] == "ZH_HToUps1SG":
    effective_xsec = higgs_ZH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "ZH_HToUps2SG":
    effective_xsec = higgs_ZH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "ZH_HToUps3SG":
    effective_xsec = higgs_ZH_xsec*higgs_3S_br*upsilon_3S_to_mumu

  # Higgs - VBFH
  if dataset["dataset_name"] == "VBFH_HToUps1SG":
    effective_xsec = higgs_VBFH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "VBFH_HToUps2SG":
    effective_xsec = higgs_VBFH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "VBFH_HToUps3SG":
    effective_xsec = higgs_VBFH_xsec*higgs_3S_br*upsilon_3S_to_mumu

  # Higgs - ttH
  if dataset["dataset_name"] == "ttH_HToUps1SG":
    effective_xsec = higgs_ttH_xsec*higgs_1S_br*upsilon_1S_to_mumu

  if dataset["dataset_name"] == "ttH_HToUps2SG":
    effective_xsec = higgs_ttH_xsec*higgs_2S_br*upsilon_2S_to_mumu

  if dataset["dataset_name"] == "ttH_HToUps3SG":
    effective_xsec = higgs_ttH_xsec*higgs_3S_br*upsilon_3S_to_mumu


  # peaking Backgrounds
  if dataset["dataset_name"] == "ZGTo2MuG_MMuMu-2To15":
    effective_xsec = z_total_xsec*mumugamma_fsr_br

  if dataset["dataset_name"] == "GluGluHToMuMuG_M125_MLL-0To60_Dalitz_012j":
    effective_xsec = higgs_total_xsec*dalitz_decay_br

  # return effective x-section
  return effective_xsec


def get_MC_weight(dataset, year, sign_weighted_total_of_events):
  # Data
  if dataset["dataset_name"].startswith("Run"): 
    return 1.0

  # get integrated lumi
  lumi = None
  if year == "2016":
    lumi = 35.9*1E3
  if year == "2017":
    lumi = 27.13*1E3
  if year == "2018":
    lumi = 59.74*1E3

  # get cross section for the process
  xsec = get_xsec(dataset)

  event_weight = xsec*lumi/sign_weighted_total_of_events
  return event_weight
```

Creating a function to select events

```python
def selector_evaluator(dataset, cut_value1, cut_value2, cut_value3, selected_events):

  # get accumulator and load pre-selected events
  dataset_name = dataset['dataset_name']
  year = dataset["year"]

  print("--> Starting processing of: "+ dataset["dataset_name"] + " - "+ dataset["year"])

  df = load("input_files/sel_" + dataset_name + "_" + year + ".coffea")

  weight = 1.0

  if not(dataset["dataset_name"].startswith("Run")): # if it is MC
    cutflow = float(df["cutflow"]["all_events"])
    weight = get_MC_weight(d, year, cutflow)


  n_evt_cut_E1 = [0 for v in cut_value1] 
  n_evt_cut_E2 = [0 for v in cut_value1] 
  n_evt_cut_F1 = [0 for v in cut_value1] 
  n_evt_cut_F2 = [0 for v in cut_value2] 
  n_evt_cut_H1 = [0 for v in cut_value3] 
  n_evt_cut_H2 = [0 for v in cut_value3] 
    
  selected_events = {
    "dataset_name" : dataset_name,
    "year": year,
    "all_events": df["cutflow"]["all_events"],
    "events_cut_E1": n_evt_cut_E1,
    "events_cut_E2": n_evt_cut_E2,
    "events_cut_F1": n_evt_cut_F1,
    "events_cut_F2": n_evt_cut_F2,
    "events_cut_H1": n_evt_cut_H1,
    "events_cut_H2": n_evt_cut_H2,
  }
    
  #loop over all cuts values  
  for i_cut, cut in enumerate(cut_value1):
    #loop over all events
    for evt in range(df["leadmu_pt"].value.size):
        
      if df["delta_R_leading_photon"].value[evt] >= cut:
        selected_events["events_cut_E1"][i_cut] += 1.0*weight
        
      if df["delta_R_trailing_photon"].value[evt] >= cut:
        selected_events["events_cut_E2"][i_cut] += 1.0*weight
        
      if df["delta_R_upsilon_photon"].value[evt] >=  cut:
        selected_events["events_cut_F1"][i_cut] += 1.0*weight  
        
  #loop over all cuts values  
  for i_cut, cut in enumerate(cut_value2):
    #loop over all events
    for evt in range(df["leadmu_pt"].value.size):    
    
      if math.fabs(df["delta_Phi_leading_photon"].value[evt]) >= cut:
        selected_events["events_cut_F2"][i_cut] += 1.0*weight
        
  #loop over all cuts values  
  for i_cut, cut in enumerate(cut_value3):
    #loop over all events
    for evt in range(df["leadmu_pt"].value.size):
        
      if df["et_photon_over_M_boson"].value[evt] >= cut:
        selected_events["events_cut_H1"][i_cut] += 1.0*weight  
        
      if df["pt_upsilon_over_M_boson"].value[evt] >=  cut:
        selected_events["events_cut_H2"][i_cut] += 1.0*weight 

      
  return selected_events
```

Function that calculates the Figure of Merit

```python
def fom(sel_mc_evt, sel_data_evt):
    return np.divide(sel_mc_evt, np.power(sel_data_evt, 0.5))
```
